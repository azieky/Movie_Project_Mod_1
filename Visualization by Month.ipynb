```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('final_df.csv')
```

```python
df.head()
```

```python
#Drop Outlier 'The Gallows'
df['dom_roi'].sort_values(ascending=False)[:10]
df.iloc[1691]
df = df.drop([1691])
```

```python
# Sort Domestic Gross by Month
df.groupby(['release_month'])['domestic_gross'].mean().sort_values(ascending=False)
```

```python
#Plot Domestic Gross by Month
Months = ['Sep', 'Oct', 'Jan', 'Apr', 'Aug', 'Feb','Mar', 'Nov', 'Jul', 'May', 'Jun', 'Dec']
x = Months
y = df.groupby(['release_month'])['domestic_gross'].mean().sort_values(ascending=True)
ax = sns.barplot(x, y, data=df)

```

```python
# Sort Worldwide Gross by Month
df.groupby(['release_month'])['worldwide_gross'].mean().sort_values(ascending=False)
```

```python
#Plot Worldwide Gross by Month

#Build Plot
plt.figure(figsize =(10,6)) 
plt.title('Worldwide Gross by Release Month')
plt.xlabel('Release Month')
plt.ylabel('Worldwide ROI by 10 Million')

#Customize Seaborn
sns.set_style("darkgrid")
sns.set_context("paper", rc={"font.size":25,"axes.titlesize":25,"axes.labelsize":15})

#Add Data
Months_2 = ['Sep', 'Jan', 'Oct', 'Aug', 'Apr', 'Feb','Mar', 'Nov', 'Dec', 'Jul', 'Jun', 'May']
x_2 = Months_2
y_2 = df.groupby(['release_month'])['worldwide_gross'].mean().sort_values(ascending=True)
ax_2 = sns.barplot(x_2, y_2, data=df)
```

```python
#Sort Domestic Return of Investment by Month
df.groupby(['release_month'])['dom_roi'].mean().sort_values(ascending=False)
```

```python
#Plot Domestic Return of Investment by Month

#Build Plot
plt.figure(figsize =(10,6)) 
plt.title('Domestic Return of Investment by Release Month')
plt.xlabel('Release Month')
plt.ylabel('Domestic ROI by %')

#Customize Seaborn
sns.set_style("darkgrid")
sns.set_context("paper", rc={"font.size":25,"axes.titlesize":25,"axes.labelsize":15})

#Add Data
Months_3 = ['May', 'Dec', 'Nov', 'Sep', 'Jun', 'Apr','Aug', 'Feb', 'Oct', 'Mar', 'Jul', 'Jan']
x_3 = Months_3
y_3 = df.groupby(['release_month'])['dom_roi'].mean().sort_values(ascending=True)
ax_3 = sns.barplot(x_3, y_3, data=df)
```

```python
#Sort Worldwide Return of Investment by Month
df.groupby(['release_month'])['world_roi'].mean().sort_values(ascending=False)
```

```python
#Plot Worldwide Return of Investment by Month

#Build Plot
plt.figure(figsize =(10,6)) 
plt.title('Worldwide Return of Investment by Release Month')
plt.xlabel('Release Month')
plt.ylabel('Worldwide ROI by %')
plt.tight_layout()
plt.show

#Customize Seaborn
sns.set_style("darkgrid")
sns.set_context("paper", rc={"font.size":20,"axes.titlesize":25,"axes.labelsize":10})

# Add data
Months_4 = ['Sep', 'Mar', 'Aug', 'Apr', 'May', 'Feb','Dec', 'Nov', 'Jun', 'Oct', 'Jul', 'Jan']
x_4 = Months_4
y_4 = df.groupby(['release_month'])['world_roi'].mean().sort_values(ascending=True)
ax_4 = sns.barplot(x_4, y_4, data=df)

```
