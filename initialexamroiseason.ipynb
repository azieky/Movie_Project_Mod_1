```python
budgets = pd.read_csv('tn.movie_budgets.csv')
budgets.head()
```

```python
budgets.shape
```

```python
budgets.iloc[:,3:] = budgets.iloc[:,3:].applymap(lambda x: float(x[1::].replace(',','')))
budgets.head()
```

```python
budgets['dom_roi']=(budgets.domestic_gross-budgets.production_budget)/budgets.production_budget
```

```python
budgets['world_roi'] = (budgets.worldwide_gross - budgets.production_budget)/budgets.production_budget
```

```python
budgets.head()
```

```python
budgets['release_month']= [str(date[0].strip()) for date in budgets.release_date.apply(lambda x: x.split())]
```

```python
budgets.head()
```

```python
budgets['release_day']= [int(date[1].rstrip(',') )for date in budgets.release_date.apply(lambda x: x.split())]
budgets['release_year']= [int(date[2])for date in budgets.release_date.apply(lambda x: x.split())]
```

```python
budgets.head()
```

```python
budgets.movie
```

```python
budgets.release_year.unique()
```

```python
budgets.head()
```

```python
budgets.release_month.unique()
```

```python
winter= ['Dec', 'Jan', 'Feb']
spring=['Mar', 'Apr','May']
summer = ['Jun', 'Jul', 'Aug']
fall = ['Sep', 'Oct', 'Nov']
budgets['season'] = budgets.release_month.apply(lambda x: 'winter' if x in winter else 'spring' if x in spring else 'summer' if x in summer else 'Fall')
```
