```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#three data sets to merge
original_df = pd.read_csv('roiimdb.csv')
rating_df = pd.read_csv('imdbrough.csv')
keyword_df = pd.read_csv('keyword_df.csv')
```

```python
#add column with imdbcode to original df
original_df['imdbcode'] = original_df.imdblink.apply(lambda x : x[7:].rstrip('/'))
```

```python
original_df.head()
```

```python
#change column names
rating_df.columns = ['imdbcode', 'movie*', 'year*', 'MPAA Rating','Runtime','Genre','Release Date*', 'Origin', 'IMDB Score','1*','2*']
rating_df
```

```python
#remove unneeded columns
rating_df = rating_df.drop(columns=['movie*', 'year*', 'Release Date*','1*','2*'])
rating_df
```

```python
#merge original and rating dataframes
final_df = pd.merge(original_df,rating_df,on='imdbcode')
final_df
```

```python
#drop more unneeded columns
final_df = final_df.drop(columns=['Unnamed: 0','level_0','index','id'])
```

```python
final_df.head()
```

```python
#merge keywords into final_df
final_df = pd.merge(final_df,keyword_df,on='imdbcode')
```

```python
final_df.to_csv('final_df.csv')
```
