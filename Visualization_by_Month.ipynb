```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('final_df.csv')
```

```python
df.head(1)
```

```python
#Drop Outlier 'The Gallows'
df['dom_roi'].sort_values(ascending=False)[:10]
df.iloc[1691]
df = df.drop([1691])
```

```python
# Sort Domestic Gross by Month
df.groupby(['release_month'])['domestic_gross'].mean().sort_values(ascending=False)
```

```python
#Plot Domestic Gross by Month

#Order of Months
Months = ['Sep', 'Oct', 'Jan', 'Apr', 'Aug', 'Feb','Mar', 'Nov', 'Jul', 'May', 'Jun', 'Dec']


#Plot data
plt.figure(figsize =(10,6))
sns.set_style("darkgrid")
sns.set_context("paper", rc={"font.size":20,"axes.titlesize":25,"axes.labelsize":10})
dom_gross_month = sns.barplot(x = df['release_month'], y = df['domestic_gross'],
                     data = df,
                     order=Months,
                     linewidth=2.5,
                     errcolor=".2",edgecolor=".2")
#Labels
plt.title('Mean Domestic Gross by Release Month')
plt.xlabel('Release Month')
plt.ylabel('Domestic Gross per 10 Million')
plt.tight_layout()
plt.show()
```

```python
# Sort Worldwide Gross by Month
df.groupby(['release_month'])['worldwide_gross'].mean().sort_values(ascending=False)
```

```python
#Order Month
Months_2 = ['Sep', 'Jan', 'Oct', 'Aug', 'Apr', 'Feb','Mar', 'Nov', 'Dec', 'Jul', 'Jun', 'May']

#Plot Data
plt.figure(figsize =(10,6))
sns.set_style("darkgrid")
sns.set_context("paper", rc={"font.size":20,"axes.titlesize":25,"axes.labelsize":10})
world_gross_month = sns.barplot(x = df['release_month'], y = df['worldwide_gross'],
                     data = df,
                     order=Months_2,
                     linewidth=2.5,
                     errcolor=".2",edgecolor=".2")

#Labels
plt.title('Mean Worldwide Gross by Release Month')
plt.xlabel('Release Month')
plt.ylabel('World Gross per 10 Million')
plt.tight_layout()
plt.show()
```

```python
#Sort Domestic Return of Investment by Month
df.groupby(['release_month'])['dom_roi'].mean().sort_values(ascending=False)
```

```python
#Plot Domestic Return of Investment by Month
Months_3 = ['May', 'Dec', 'Nov', 'Sep', 'Jun', 'Apr','Aug', 'Feb', 'Oct', 'Mar', 'Jul', 'Jan']

#Build Figure
plt.figure(figsize =(10,6))
sns.set_style("darkgrid")
sns.set_context("paper", rc={"font.size":20,"axes.titlesize":25,"axes.labelsize":10})
dom_roi_month = sns.barplot(x = df['release_month'], y = df['dom_roi'],
                     data = df,
                     order=Months_3,
                     linewidth=2.5,
                     errcolor=".2",edgecolor=".2")
#Labels
plt.title('Mean Domestic Return on Investment by Release Month')
plt.xlabel('Release Month')
plt.ylabel('Return on Investment')
plt.tight_layout()
plt.show()
```

```python
#Sort Worldwide Return of Investment by Month
df.groupby(['release_month'])['world_roi'].mean().sort_values(ascending=False)
```
